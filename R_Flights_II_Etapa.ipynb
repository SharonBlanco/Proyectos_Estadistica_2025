**Comparación de Varianzas de Distancias de Vuelo entre Aeropuertos JFK y LGA**


**PROBLEMA**

Con el objetivo de mejorar la eficiencia en la asignación de aeronaves y recursos, el equipo de logística de una aerolínea desea comparar la variabilidad en las distancias de vuelo de los aeropuertos JFK (John F. Kennedy) y LGA (LaGuardia).

Según informes previos, se sospecha que la variabilidad de la distancia de los vuelos de cada aeropuerto es similar. Por lo tanto, se desea realizar una prueba de hipótesis para determinar si se puede suponer que la varianza de las distancias de los vuelos de JFK es igual a la de LGA, con un nivel de significancia del 5%.

Se asume que ambas poblaciones siguen una distribución normal, y se toma JFK como
𝑚
 y LGA como
𝑛
.


Sabiendo que:


 $H_0: \frac{\sigma_m^2}{\sigma_n^2} = 1$


 $H_1: \frac{\sigma_m^2}{\sigma_n^2} \neq  1$



**Tipo de prueba:** Dos colas de distribución F.



Para validar dicha afirmación, se procede a realizar la prueba en R:

Para investigar esta diferencia, se recopilan dos muestras aleatorias de 20 vuelos con reemplazo desde una base de datos, seleccionando la variable distance de la base de datos flights del paquete nycflights13, que representa la distancia recorrida por cada vuelo en millas.





```r
# 1. Fijar semilla para reproducibilidad
set.seed(1000)

# 2. Filtrar vuelos según el origen JFK y LGA
JFK.flights <- subset(flights, origin == "JFK")   # Todos los vuelos desde JFK
LGA.flights <- subset(flights, origin == "LGA")   # Todos los vuelos desde LGA

# 3. Tomar una muestra aleatoria de 20 vuelos de cada uno con reemplazo
sample.JFK <- JFK.flights[sample(1:nrow(JFK.flights), 20, replace = T), ]
sample.LGA <- LGA.flights[sample(1:nrow(LGA.flights), 20, replace = T), ]

#4. Calcular el resultado
resultado.varianzas <- stests::var.test(sample.JFK$distance, sample.LGA$distance, alternative = "two.sided", conf.level= 0.95)

# 5. Se realiza la prueba de hipótesis
cat("===== Prueba de hipótesis del COCIENTE DE VARIANZAS de los vuelos =====\n")
print(resultado.varianzas)
```

# **Prueba de hipotesis para 2 poblaciones utilizando T-Student**

**Problema:**

Con el fin de optimizar la asignación de aeronaves según la duración promedio de los vuelos, el equipo de planificación de rutas de una aerolínea desea investigar si existe una diferencia significativa en la distancia promedio de los vuelos que parten del aeropuerto JFK (John F. Kennedy) respecto a los del LGA (LaGuardia), ambos en la ciudad de Nueva York.

Aunque se sabe que los destinos de ambos aeropuertos pueden variar, se sospecha que las distancias promedio son similares, lo cual permitiría asignar los mismos tipos de aeronaves a ambos aeropuertos. Para confirmar esta suposición, se decide realizar una prueba de hipótesis con un nivel de significancia del 5%.

**Sabiendo que:**

* μ₁ = media poblacional de las distancias desde JFK

* μ₂ = media poblacional de las distancias desde LGA

* H₀: μ₁ = μ₂ (las distancias promedio desde JFK y LGA son iguales).

* H₁: μ₁ ≠ μ₂ (las distancias promedio desde JFK y LGA son diferentes).
* Las dos muestras son independientes.

* Las poblaciones tienen distribución normal.

* Según el analisis de conciente de varianzas realizado previamente, se puede asumir que las varianzas son iguales (p > 0.05).

```r
# 1. Prueba t para diferencia de medias, asumiendo varianzas iguales
resultado.medias <- t.test(sample.JFK$distance, sample.LGA$distance,
                           var.equal = TRUE, alternative = "two.sided", conf.level = 0.95)

# 2. Mostrar resultado
cat("===== Prueba T para Diferencia de Medias entre JFK y LGA =====\n")
print(resultado.medias)
```

### **Resumen de prueba:**

* Significancia estadística	α =  0.05
* Nivel de confianza = 0.95
* Valor P de la prueba = 0.9515

**Conclusión:**

Dado que el valor p = 0.9515 es mucho mayor que el nivel de significancia α = 0.05, no existe evidencia estadística suficiente para rechazar la hipótesis nula H₀. Es decir, no se puede afirmar que haya una diferencia significativa entre la distancia promedio de los vuelos que parten desde los aeropuertos JFK y LGA.

## **Promedio de Retraso de Salida en el Aeropuerto LGA Usando Distribución Z**
### **Problema**

El equipo de operaciones del aeropuerto LaGuardia (LGA) quiere evaluar si el promedio de retraso en salidas de sus vuelos supera los 10 minutos, lo cual es el estándar considerado aceptable por la gerencia para una operación eficiente.

Se supone que el retraso en salida tiene una distribución normal y se conoce la desviación estándar muestral. Para verificar, se toma una muestra aleatoria de vuelos y se aplica una prueba Z para el promedio, con un nivel de confianza del 95%. Se desea hacer una prueba de hipótesis para verificar lo dicho por la compañía.

Sabiendo que:

H₀: μ = 10 minutos (≤)

H₁: μ > 10 minutos

**Tipo de prueba:** Z de cola derecha (unilateral superior)

Para validar dicha afirmación, se procede a realizar la prueba en R:

Para investigar el retraso promedio de las salidas de vuelos desde el aeropuerto LGA, se recopila una muestra aleatoria de 40 vuelos con reemplazo desde una base de datos, seleccionando la variable dep_delay, que representa el retraso de salida en minutos. Bajo el supuesto de normalidad en la distribución de los datos y con una desviación estándar muestral conocida, se aplica una prueba Z para una media con un nivel de confianza del 95%.






```r
#Instalación del paquete stests
if (!require('devtools')) install.packages('devtools')
devtools::install_github('fhernanb/stests', force=TRUE)
```

```r
set.seed(1000)

# Filtrar vuelos del aeropuerto LGA y eliminar NA en dep_delay
LGA.flights <- subset(flights, origin == "LGA" & !is.na(dep_delay))

# Tomar una muestra aleatoria de 40 vuelos con reemplazo
sample.LGA <- LGA.flights[sample(1:nrow(LGA.flights), 40, replace = TRUE), ]

# Calcular la desviación estándar muestral
std_dev_dep_delay <- sd(sample.LGA$dep_delay)
# Aplicar la prueba Z
resultado <- stests::z.test(
  x = sample.LGA$dep_delay,
  sigma2 = std_dev_dep_delay,
  mu = 10,
  conf.level = 0.95,
  alternative = "greater"
)

cat("===== Prueba de hipótesis de 95% para el Promedio de Retraso =====\n")
print(resultado)
```

# **Prueba de Hipótesis para dos poblaciones con la distribución z**

**Problema:**

Los meteorólogos quieren saber si la temperatura promedio en julio supera significativamente a la de enero en Nueva York. Se asume que las temperaturas de ambos meses siguen una distribución normal. Para ello, se toma una muestra aleatoria de datos de temperatura diaria de cada mes y, con un nivel de significancia del 5 %, se realiza una prueba de hipótesis de dos muestras independientes.

**Hipótesis:**

Sabiendo que:

H₀: μ_Julio – μ_Enero = 0

H₁: μ_Julio – μ_Enero > 0

**Tipo de prueba:**

Cola derecha(temperatura de Julio mayor que la de Enero)
Nivel de significancia: α = 0.05

**Distribución:**

Z (normal estándar)

**Condiciones asumidas:**

Muestras aleatorias e independientes.

La variable “temperatura” sigue aproximadamente una distribución normal en cada mes.

Desviaciones estándar muestrales conocidas.


```r
install.packages("PASWR2")
library(PASWR2)
```

```r
# 1. Fijar semilla para reproducibilidad
set.seed(2025)

# 2. Filtrar datos de temperaturas para Julio (mes 7) y Enero (mes 1)
data_jul <- subset(weather, month == 7 & !is.na(temp))$temp
data_jan <- subset(weather, month == 1 & !is.na(temp))$temp

# 3. Tomar muestra aleatoria de 30 observaciones con reemplazo
sample_jul <- sample(data_jul, 30, replace = TRUE)
sample_jan <- sample(data_jan, 30, replace = TRUE)

# 4. Estimamos las desviaciones
sigma_jul <- sd(sample_jul)
sigma_jan <- sd(sample_jan)

# 5. Prueba Z (greater)
z_result <- PASWR2::z.test(x = sample_jul,
                   y = sample_jan,
                   sigma.x = sigma_jul,
                   sigma.y = sigma_jan,
                   alternative = "greater",
                   mu = 0,
                   conf.level = 0.95)

# 6. Mostrar resumen
print("===== Resultados de la prueba Z (Julio > Enero) =====")
print(z_result)
```

**Resumen de la prueba:**

Valor observado: $Z_{\text{obs}}$ = 19.462

**Análisis:**
Para esta prueba se comparó la temperatura diaria promedio de julio frente a la de enero en Nueva York, usando muestras independientes de $n=30$ observaciones (con reemplazo) en cada mes. Se generaron así dos vectores de datos (`sample_jul` y `sample_jan`) tras fijar semilla para reproducibilidad.

La media de julio resultó $\bar x_{Julio}=80.288$ y la de enero $\bar x_{Enero}=33.374$, con una diferencia muestral de $46.914$. Las desviaciones estándar muestrales se estimaron como $\sigma_{Julio}=sd(sample\_jul)$ y $\sigma_{Enero}=sd(sample\_jan)$. A partir de ellas se calculó el error estándar

$$
  SE = \sqrt{\frac{\sigma_{Julio}^2}{30} + \frac{\sigma_{Enero}^2}{30}}
$$

y el estadístico

$$
  Z_{\rm obs}
  = \frac{(\bar x_{Julio}-\bar x_{Enero}) - 0}{SE}
  = 19.462.
$$

Dado que la diferencia observada $46.914$ cae en la región de rechazo $\bigl]42.94911,\infty\bigr[$ y el p‑valor (\~$2.2\times10^{-16}$) es muchísimo menor que 0.05, **rechazamos la hipótesis nula**.
